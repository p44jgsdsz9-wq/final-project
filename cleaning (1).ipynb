```python
import pandas as pd
import numpy as np


# Load dataset
def load_data(path):
    df = pd.read_csv("/content/AmesHousing.csv")
    return df


# Explore basic information
def explore_data(df):

    print("First 5 rows:")
    print(df.head())

    print("\nShape:", df.shape)

    print("\nColumn names:")
    print(df.columns)

    print("\nData types:")
    print(df.dtypes)


# Detect missing values
def check_missing(df):

    missing = df.isnull().sum().sort_values(ascending=False)

    print("\nMissing values:")
    print(missing)

    return missing


# Detect duplicates
def check_duplicates(df):

    duplicates = df.duplicated().sum()

    print("\nDuplicates:", duplicates)

    return duplicates


# Main cleaning function
def clean_data(df):

    # Fix incorrect data types
    if "MSSubClass" in df.columns:
        df["MSSubClass"] = df["MSSubClass"].astype(str)

    # Drop columns with more than 50% missing values
    df = df.drop(columns=df.columns[df.isna().mean() > 0.5])

    # Fill numeric columns with median
    num_cols = df.select_dtypes(include=np.number).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())

    # Fill categorical columns with "Unknown"
    cat_cols = df.select_dtypes(include="object").columns
    df[cat_cols] = df[cat_cols].fillna("Unknown")

    # Remove duplicates
    df = df.drop_duplicates()

    # Outlier capping for SalePrice
    if "SalePrice" in df.columns:
        upper = df["SalePrice"].quantile(0.99)
        df["SalePrice"] = df["SalePrice"].clip(upper=upper)

    return df


# Final checks
def final_checks(df):

    print("\nTotal missing values:", df.isnull().sum().sum())

    if "SalePrice" in df.columns:
        print("All SalePrice > 0:", (df["SalePrice"] > 0).all())

    print("Final shape:", df.shape)


# Run pipeline
df = load_data("AmesHousing.csv")

explore_data(df)

check_missing(df)

check_duplicates(df)

df = clean_data(df)

final_checks(df)
df.to_csv("clean_data.csv",index=False)
```
